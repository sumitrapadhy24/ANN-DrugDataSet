```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
from sklearn import datasets
```

```python
x,y=datasets.make_moons(n_samples=1000)
```

```python
print(x.shape)
print(y.shape)
```

```python
sns.scatterplot(x[:,0],x[:,1],hue=y)
plt.show()
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
algo=MLPClassifier(hidden_layer_sizes=(60,40),activation='relu',max_iter=200)
```

```python
algo.fit(x,y)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x,y,algo)
plt.show()
```
